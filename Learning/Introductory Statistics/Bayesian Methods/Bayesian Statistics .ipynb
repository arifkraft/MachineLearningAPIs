# Thinking Bayesian 
The idea behind Bayesian approach is more intuitive than a frequentist approach. However, its application is limited in the field of statistics. To illustrate the difference let us consider a simple example -
<br>
We would like to ascertain if Steve is a good coder. Looking at his performance, out of 10 instances where he had written a script, 8 were error-free. A frequentist would say, probability that he's a good coder is - <br><br>
$$ \frac{8}{10} = 0.8 $$ <br>
However, it may not be a good thing to tell tell the likelihood of him being a good coder just by looking at 10 instances because generally only 10% coders write error-free codes in the first go (*Prior Belief*). Here comes the bayesian approach. <br>
The likelihood of Steve being a good coder can be obtained using the Baye's theorem - <br><br>
$$ P(A|X) = \frac{P(X|A).P(A)}{P(X)} $$ <br>
* A is an event which tells Steve is a good coder
* X is the evidence (8 out of 10 scripts are error-free)
<br><br>
So,
1. P(A|X) : Probability of Steve being a good coder, given the evidence (8 out of 10 scripts are error-free)
    * *We need to find this*
1. P(X|A) : How likely is it that Steve would write 8 out of 10 error-free codes, given he is infact a good coder 
    * *This should be high. Lets say 0.8 (another)*  
1. P(A) : Probability of writing an error-free code in general (Prior Probability without any evidence)
    * *Lets say 0.1*
1. P(X) : This one is tricky. How likely is it that Steve would write 8 out of 10 codes without an error. No prior knowledge of whether he's a good coder or not. Let us do some refactoring to obtain this
    * $$ P(X) = P(X\space and\space A) \space OR \space P(X\space and\space ~A) $$<br>
    $$  P(X) = P(X|A).P(A) \space + \space P(X|~A).P(~A)$$<br>
    $$  P(X) = 0.8x0.1 \space + \space 0.2x0.9 $$<br>
    $$  P(X) = 0.28 $$
1. Applying the Baye's Theorem -
    $$ P(A|X) = \frac{P(X|A).P(A)}{P(X)} $$<br>
    $$ P(A|X) = \frac{0.8 x 0.1}{0.28}$$<br>
    $$ P(A|X) = 0.285 $$<br>
    
    

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt

figsize = [7.5,2]
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300

prior = [1/10, 1-(1/10)]
posterior = [0.285, 1-0.28]

plt.bar([0,0.7], prior, alpha =0.7, width = 0.25,
       label = 'Prior Distribution', lw ='3')

plt.bar([0+0.25, 0.7+0.25], posterior, alpha =0.7, width = 0.25,
       label ='Posterior Distribution', lw='3')
plt.xticks([0.125,0.825],["Good Coder","Bad Coder"])
plt.title('Steve\'s Prior and Posterior probabilities')

plt.ylabel('Probability')
plt.legend(loc= 'upper left')
```
